Image Processing with SciPy and NumPy

```python
import matplotlib.pyplot as plt

img = io.imread(r'/content/image.jpg')
king=img.copy()
plt.imshow(img)
plt.axis('off')
plt.title("Kingfisher")
plt.show()
```

```python
img = np.fromfile('king.raw', dtype=np.uint8)
print(img.shape)
```

```python
x, y, _ = king.shape
crop = king[x//3: -x//8, y//3: -y//8]
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Cropped Image")
plt.show()
```

```python
flipped = np.flipud(king)
plt.imshow(flipped)
plt.axis('off')
plt.title("Flipped Image (Up-Down)")
plt.show()
```

```python
from scipy.ndimage import gaussian_filter
blurred = gaussian_filter(king, sigma=5)

plt.imshow(blurred.astype(np.uint8))
plt.axis('off')
plt.title("Gaussian Blurred")
plt.show()
```
